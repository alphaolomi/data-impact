```python
data['total']=data.sum(axis=1)
data.total
```

```python
## first 10 councils with large number of incidences
data.nlargest(10, data[['total']], keep='first')
```

```python
## first 10 regions with less number of incidences
data.nsmallest(10, data[['total']], keep='first')
```

### First Ten Diseases with large Occurance

```python
### ftd means "first ten diseases with large occurance".
ftd=data.mean().sort_values(ascending=False)[0:11][1:]
ftd
```

```python
ftd.plot(kind="bar")
plt.title("Top ten disease in 2015 for \n under 5 IPD")
plt.ylabel("Number of inpatient department diagnoses (IPD)")
plt.savefig("../figure/2015_top_ten_disease.png", bbox_inches="tight", dpi=1200)
```
